## EXIT codes in spell_clean.csv

### Good:
- XRF: reunified with family
- XLC: legal custodianship
- XRL: to relative (obsolete code)
- XCA: adopted

### Bad:
- XOT: other
- XOP: other (obsolete code)
- XRM: age out
- XRY: runaway
- XJP: juvenile probation

### Unknown:
- ZTC: spell ongoing
- XUK: unknown

```python
good_val = 'GOOD'
bad_val = 'BAD'
unknown_val = 'UNKNOWN'
exit_mapping = {
    'XRF':good_val, 'XLC':good_val, 'XRL':good_val, 'XCA':good_val,
    'XOT':bad_val, 'XOP':bad_val, 'XRM':bad_val, 'XRY':bad_val, 'XJP':bad_val,
    'ZTC':unknown_val, 'XUK':unknown_val
}
outcomes = [exit_mapping[x] for x in spells.EXIT if x in exit_mapping]
sns.countplot(outcomes)
spells['outcome'] = outcomes
```

```python
# sns.countplot(x = 'NPLACES', hue = 'outcome', data=spells)
nplaces_pmf_good = lv.get_pmf(spells[spells.outcome=='GOOD'].NPLACES)
nplaces_pmf_bad = lv.get_pmf(spells[spells.outcome=='BAD'].NPLACES)
# sns.barplot(nplaces_pmf_good.index, nplaces_pmf_good)
g = sns.FacetGrid(spells, col='outcome')
g.map(sns.distplot, 'NPLACES', kde=False);
```
